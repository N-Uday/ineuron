## Question - 1

```python
'Q1. '

class ListNode(object):
    def __init__(self, x):
        self.val = x
        self.next = None
        
class Solution(object):
    def greaterNodes(self, head1, head2):
        
        # Checking for empty values
        if not head1 or not head2:
            return None
        
        # Assigning pointers for both linked lists
        p, q = head1, head2
        # Assuming head & tail of new linked list as null
        head3, tail = None, None
        
        # Loop ends when p becomes null
        while p:
            # Creating new linked list node with the max value out of both the linked lists
            new_node = ListNode(max(p.val, q.val))
            p = p.next
            q = q.next
            
            # Checking if head3 is null means for first node
            if not head3:
                head3 = new_node
                tail = new_node
            # If not this is from Second node
            else:
                tail.next = new_node
                tail = new_node
        
        return head3    
    
##########################################################################################################################  

# Create Linked List
def CreateLL(nums):
    head = ListNode(nums[0])
    curr = head
    for i in nums[1:]:
        temp = ListNode(i)
        curr.next = temp
        curr = temp
    return head   
    
# Print Linked List
def printLL(head, string):
    curr = head
    print(string,' : ', end = ' ')
    while curr:
        print(curr.val, end = " -> ")
        curr = curr.next
    print(None)
    
# Declaring values and checking for output

obj = Solution()

# Example - 1

nums1 = [5, 2, 3, 8]
nums2 = [1, 7, 4, 5]
head1 = CreateLL(nums1)
head2 = CreateLL(nums2)
result = obj.greaterNodes(head1, head2)
printLL(result, 'Example - 1')


# Example - 2

nums1 = [2, 8, 9, 3]
nums2 = [5, 3, 6, 4]
head1 = CreateLL(nums1)
head2 = CreateLL(nums2)
result = obj.greaterNodes(head1, head2)
printLL(result, 'Example - 2')

# Time Complexity  - O(n)
# Space Complexity - O(1)
```
