```python
'Q1. '

class ListNode(object):
    def __init__(self, x):
        self.val = x
        self.next = None
        
class Solution(object):
    def deleteMiddle(self, head):
       
        if not head.next:
            return None

        slow = head
        fast = head
        
        
        while fast and fast.next:
            prev = slow
            slow = slow.next
            fast = fast.next.next

        prev.next = slow.next

        return head
    
##########################################################################################################################  

# Create Linked List
def CreateLL(nums):
    head = ListNode(nums[0])
    curr = head
    for i in nums[1:]:
        temp = ListNode(i)
        curr.next = temp
        curr = temp
    return head   
    
# Print Linked List
def printLL(head, string):
    curr = head
    print(string,' : ', end = ' ')
    while curr:
        print(curr.val, end = "->")
        curr = curr.next
    print(None)
    
# Declaring values and checking for output

obj = Solution()

# Example - 1

nums1 = [1,2,3,4,5]
head = CreateLL(nums1)
result = obj.deleteMiddle(head)
printLL(result, 'Example - 1')


# Example - 2

nums2 = [2, 4, 6, 7, 5, 1]
head = CreateLL(nums2)
result = obj.deleteMiddle(head)
printLL(result, 'Example - 2')

# Time Complexity  - O(n)
# Space Complexity - O(1)
```
