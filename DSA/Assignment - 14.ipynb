## Question - 1

```python
'Q1. '

class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
class Solution(object):
    def RemoveLoop(self, head, X):
        if X == 0:
            return head
        slow, fast = head, head
        
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            
            if slow == fast:
                break
                
        else:return head
        
        prev = head
        
        while slow.next != prev:
            prev = prev.next
            slow = slow.next
            
        slow.next = None
        
        return head
    
##########################################################################################################################  

# Create Linked List
def CreateLL(nums, X):
    head = ListNode(nums[0])
    curr = head
    for i in nums[1:]:
        temp = ListNode(i)
        curr.next = temp
        curr = temp
    if X > 0:
        temp = head
        for i in range(X):
            temp = temp.next
        curr.next = temp
        
    return head   
    
# Print Linked List
def printLL(head, string):
    curr = head
    print(string,' : ', end = ' ')
    while curr:
        print(curr.val, end = " -> ")
        curr = curr.next
    print(None)
    
# Declaring values and checking for output

obj = Solution()

# Example - 1

N1 = 3
nums1 = [1,3,4]
X1 = 2
head = CreateLL(nums1, X1)
result = obj.RemoveLoop(head, X1)
printLL(result, 'Example - 1')


# Example - 2

N2 = 4
nums2 = [1, 8, 3, 4]
X2 = 0
head = CreateLL(nums2, X2)
result = obj.RemoveLoop(head, X2)
printLL(result, 'Example - 2')

# Time Complexity  - O(n)
# Space Complexity - O(1)        
```

```python
'Q3. ' 
class ListNode:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.bottom = None

def merge(list1, list2):
    dummy = ListNode(0)
    tail = dummy

    while list1 and list2:
        if list1.data < list2.data:
            tail.bottom = list1
            list1 = list1.bottom
        else:
            tail.bottom = list2
            list2 = list2.bottom
        tail = tail.bottom

    if list1:
        tail.bottom = list1
    else:
        tail.bottom = list2

    return dummy.bottom

def flatten(head):
    if not head or not head.next:
        return head

    head.next = flatten(head.next)

    head = merge(head, head.next)

    return head

def printList(head):
    current = head
    while current:
        print(current.data, end=" ")
        current = current.bottom


# Example 1
head1 = ListNode(5)
head1.next = ListNode(10)
head1.next.next = ListNode(19)
head1.next.next.next = ListNode(28)

head1.bottom = ListNode(7)
head1.bottom.bottom = ListNode(8)
head1.bottom.bottom.bottom = ListNode(30)

head1.next.bottom = ListNode(20)

head1.next.next.bottom = ListNode(22)
head1.next.next.next.bottom = ListNode(35)

head1.next.next.bottom.next = ListNode(50)

head1.next.next.next.bottom = ListNode(40)
head1.next.next.next.bottom.bottom = ListNode(45)

head1 = flatten(head1)

print("\nExample - 1 :")
printList(head1)


# Example 2
head2 = ListNode(5)
head2.next = ListNode(10)
head2.next.next = ListNode(19)
head2.next.next.next = ListNode(28)

head2.bottom = ListNode(7)

head2.next.next.bottom = ListNode(22)

head2.next.next.bottom.next = ListNode(50)

head2.next.next.bottom.next.bottom = ListNode(30)

head2 = flatten(head2)

print("\n\nExample - 2 :")
printList(head2)
```

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None
        self.arbit = None

def cloneLinkedList(head):
    # Create a copy of each node and insert it between the original nodes
    current = head
    while current:
        new_node = Node(current.val)
        new_node.next = current.next
        current.next = new_node
        current = new_node.next

    # Set the arbitrary pointers of the copy nodes
    current = head
    while current:
        if current.arbit:
            current.next.arbit = current.arbit.next
        current = current.next.next

    # Separate the original and copy lists
    new_head = head.next
    current = head
    while current.next:
        temp = current.next
        current.next = temp.next
        current = temp

    return new_head

def printList(head):
    result = []
    while head:
        if head.arbit:
            result.append(f"{head.val}({head.arbit.val})")
        else:
            result.append(f"{head.val}(None)")
        head = head.next
    print(" -> ".join(result))

# Creating a linked list with random pointer
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.arbit = head.next.next
head.next.arbit = head
head.next.next.arbit = head.next.next.next.next
head.next.next.next.arbit = head.next.next
head.next.next.next.next.arbit = head.next

# Print the original list
print("The original linked list:")
printList(head)

# Function call
clone_head = cloneLinkedList(head)

print("The cloned linked list:")
printList(clone_head)
```

```python
'Q8.'
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution(object):
    def removeZeroSum(self,head):
        dummy = ListNode(0)
        dummy.next = head
    
        # Create a dictionary to store cumulative sums
        cum_sum = {}
        curr = dummy
        prefix_sum = 0
    
        # Traverse the linked list
        while curr:
            prefix_sum += curr.val
    
            if prefix_sum in cum_sum:
                # Remove the sublist from the linked list
                prev = cum_sum[prefix_sum].next
                remove_sum = prefix_sum + prev.val
    
                while prev != curr:
                    del cum_sum[remove_sum]
                    prev = prev.next
                    remove_sum += prev.val
    
                # Update the next pointer of the previous node
                cum_sum[prefix_sum].next = curr.next
    
            else:
                cum_sum[prefix_sum] = curr
    
            curr = curr.next
    
        return dummy.next
        
##########################################################################################################################  

# Create Linked List
def CreateLL(nums):
    head = ListNode(nums[0])
    curr = head
    for i in nums[1:]:
        temp = ListNode(i)
        curr.next = temp
        curr = temp
        
    return head   
    
# Print Linked List
def printLL(head, string):
    curr = head
    print(string,' : ', end = ' ')
    while curr:
        print(curr.val, end = " -> ")
        curr = curr.next
    print(None)
    
# Declaring values and checking for output

obj = Solution()

# Example - 1

nums1 = [1,2,-3,3,1]
head1 = CreateLL(nums1)
result = obj.removeZeroSum(head1)
printLL(result, 'Example - 1')


# Example - 2

nums2 = [1,2,3,-3,4]
head2 = CreateLL(nums2)
result = obj.removeZeroSum(head2)
printLL(result, 'Example - 2')

# Time Complexity  - O(n)
# Space Complexity - O(1)    
```
