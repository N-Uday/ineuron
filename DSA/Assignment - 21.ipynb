## Question - 1

```python
'Q1. '

class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def inorder_traversal(root, values):
    if root is None:
        return

    inorder_traversal(root.left, values)
    values.append(root.val)
    inorder_traversal(root.right, values)

def construct_bst(root, values):
    if not root:
        return None

    construct_bst(root.left, values)
    root.val = values[0]
    values.pop(0)
    construct_bst(root.right, values)


def binary_tree_to_bst(root):
    values = []
    inorder_traversal(root, values)
    values.sort()
    construct_bst(root, values)

# For printing
def inorder_print(root):
    if root is None:
        return

    inorder_print(root.left)
    print(root.val, end=' ')
    inorder_print(root.right)
    
# Declaring values and checking for output

# Binary Tree
root = TreeNode(10)
root.left = TreeNode(2)
root.right = TreeNode(7)
root.left.left = TreeNode(8)
root.left.right = TreeNode(4)

# Convert binary tree to BST
binary_tree_to_bst(root)
inorder_print(root)

# Time Complexity - O(NlogN)
# Space Complextiy - O(N)
```

## Question - 2

```python
'Q2. '

class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def findLCA(root, node1, node2):
    if root.val > node1.val and root.val > node2.val:
        return findLCA(root.left, node1, node2)
    elif root.val < node1.val and root.val < node2.val:
        return findLCA(root.right, node1, node2)
    else:
        return root

def findDistance(root, node, distance):
    if root.val == node.val:
        return distance
    elif root.val > node.val:
        return findDistance(root.left, node, distance + 1)
    else:
        return findDistance(root.right, node, distance + 1)

def findNodeDistance(values, node1_val, node2_val):
    # Create the binary search tree
    root = None
    for value in values:
        if root is None:
            root = TreeNode(value)
        else:
            current = root
            while True:
                if value < current.val:
                    if current.left is None:
                        current.left = TreeNode(value)
                        break
                    else:
                        current = current.left
                else:
                    if current.right is None:
                        current.right = TreeNode(value)
                        break
                    else:
                        current = current.right

    # Find the nodes
    node1 = TreeNode(node1_val)
    node2 = TreeNode(node2_val)

    # Find the lowest common ancestor
    lca = findLCA(root, node1, node2)

    # Find the distances of node1 and node2 from the LCA
    distance1 = findDistance(lca, node1, 0)
    distance2 = findDistance(lca, node2, 0)

    # Calculate the total distance
    distance = distance1 + distance2

    return distance

# Declaring values and checking for output

# Example - 1

values = [8, 3, 1, 6, 4, 7, 10, 14, 13]
node1 = 6
node2 = 14
distance1 = findNodeDistance(values, node1, node2)
print("The distance between the two keys of Example - 1 is ", distance1)

# Example - 2

values = [8, 3, 1, 6, 4, 7, 10, 14, 13]
node11 = 3
node22 = 4
distance2 = findNodeDistance(values, node11, node22)
print("The distance between the two keys of Example - 2 is ", distance2)
```
